```python
def draw_ranking_10(a, b, title, ax=None):
    alphabet = 'adefhjbcgi'
    c = [letter for letter in alphabet]
    linked_data = list(zip(a, b,c))
    linked_data.sort(key=lambda x: x[1],reverse = True)
    a, b, c = zip(*linked_data)  

    a = np.array(a)
    b = np.array(b)
    b = np.array([round(score,2) for score in b])

    cmap = colors.ListedColormap(['blue', 'red'])
    if ax is None:
        fig, ax = plt.subplots()
        ax.set_xticks([])
        ax.set_yticks([])
    ax.imshow(np.expand_dims(a, axis=1), cmap=cmap)
    ax.set_title(title)

    for i in range(a.shape[0]+1):
        ax.axhline(i-0.5, color='black', linewidth=1, linestyle='-', alpha=0.5)
    for i in range(a.shape[0]):
        ax.text(0, i, str(b[i]), fontsize=12, ha='center', va='center', color='white')
        ax.text(-1.3, i, str(c[i]), fontsize=12, ha='left', va='center', color='black')
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
fig, axs = plt.subplots(1, 3)

# Call the draw_ranking_10() function for each subplot
draw_ranking_10(classes, original, 'Original', ax=axs[0])
draw_ranking_10(classes, x0, 'Accounting for bias', ax=axs[2])
draw_ranking_10(classes, estim, 'Normal algo', ax=axs[1])
plt.show()
```

```python
#algo to rank, from learning to order things paper ----------- to remove
def GreedyOrder(X, PREF):
    V = X
    pi = {}
    for v in V:
        pi[v] = sum(PREF(v, u) for u in V) - sum(PREF(u, v) for u in V)
    
    hat_rho = {}
    while V:
        t = max(V, key=lambda u: pi[u])
        hat_rho[t] = len(V)
        V.remove(t)
        for v in V:
            pi[v] += PREF(t, v) - PREF(v, t)
    return hat_rho
```

```python
# create empty dictionary to store the results
results_dict = {}

# load DataFrame into a NumPy array
arr = df_faceage.to_numpy()

# iterate over each item in the dictionary
for key, pairs in PC_faceage.items():
    
    
    # initialize counts for each combination of genders
    num_wrong_pairs = np.zeros((2, 2))
    num_correct_pairs = np.zeros((2, 2))

    # iterate over each pair in the list
    for pair in pairs:
        i,j = pair

        score1, gender1 = arr[i, [1, 2]]
        score2, gender2 = arr[j, [1, 2]]

        # update counts based on gender and score comparisons
        if gender1 == 0:
            if gender2 == 0:
                num_correct_pairs[0, 0] += (score1 > score2)
                num_wrong_pairs[0, 0] += (score1 < score2)
            else:
                num_correct_pairs[0, 1] += (score1 > score2)
                num_wrong_pairs[0, 1] += (score1 < score2)
        else:
            if gender2 == 0:
                num_correct_pairs[1, 0] += (score1 > score2)
                num_wrong_pairs[1, 0] += (score1 < score2)
            else:
                num_correct_pairs[1, 1] += (score1 > score2)
                num_wrong_pairs[1, 1] += (score1 < score2)

    # add results to dictionary
    results_dict[key] = {
        'num_wrong_pairs_0_0': num_wrong_pairs[0, 0],
        'num_wrong_pairs_0_1': num_wrong_pairs[0, 1],
        'num_wrong_pairs_1_0': num_wrong_pairs[1, 0],
        'num_wrong_pairs_1_1': num_wrong_pairs[1, 1],
        'num_correct_pairs_0_0': num_correct_pairs[0, 0],
        'num_correct_pairs_0_1': num_correct_pairs[0, 1],
        'num_correct_pairs_1_0': num_correct_pairs[1, 0],
        'num_correct_pairs_1_1': num_correct_pairs[1, 1]
    }

# create DataFrame from results dictionary
df_results = pd.DataFrame.from_dict(results_dict, orient='index')

# display results
print(df_results)
```

```python
df_results['annot_bias'] = annot_bias
df0 = df_results[(df_results['num_wrong_pairs_1_0'] + df_results['num_correct_pairs_0_1']) > 5]
df1 = df_results[(df_results['num_wrong_pairs_0_1'] + df_results['num_correct_pairs_1_0']) > 5]
df2 = df_results[(df_results['num_wrong_pairs_0_0'] + df_results['num_correct_pairs_0_0']) > 5]
df3 = df_results[(df_results['num_wrong_pairs_1_1'] + df_results['num_correct_pairs_1_1']) > 5]
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 12))

axs[0,0].scatter(df0['annot_bias'], df0['num_wrong_pairs_1_0'] / (df0['num_wrong_pairs_1_0'] + df0['num_correct_pairs_0_1']))
axs[0,0].set_xlabel('Annotators group bias')
axs[0,0].set_ylabel('Fraction of class 1 wrongly preferred over class 0')
axs[0,0].set_title('Ground truth 0>1')


axs[0, 1].scatter(df1['annot_bias'], df1['num_wrong_pairs_0_1'] / (df1['num_wrong_pairs_0_1'] + df1['num_correct_pairs_1_0']))
axs[0, 1].set_xlabel('Annotators group bias')
axs[0, 1].set_ylabel('Fraction of class 0 wrongly preferred over class 1')
axs[0, 1].set_title('Ground truth 1>0')

axs[1, 0].scatter(df2['annot_bias'], df2['num_wrong_pairs_0_0'] / (df2['num_wrong_pairs_0_0'] + df2['num_correct_pairs_0_0']))
axs[1, 0].set_xlabel('Annotators group bias')
axs[1, 0].set_ylabel('Fraction of errors evaluation 0-0 pairs')
axs[1, 0].set_title('0-0 pairs')

axs[1, 1].scatter(df3['annot_bias'], df3['num_wrong_pairs_1_1'] / (df3['num_wrong_pairs_1_1'] + df3['num_correct_pairs_1_1']))
axs[1, 1].set_xlabel('Annotators group bias')
axs[1, 1].set_ylabel('Fraction of errors evaluation 1-1 pairs')
axs[1, 1].set_title('1-1 pairs')

plt.show()
```
