```python
np.random.seed(414)
# Filter the dataset to include only rows with sex=1 and sex=0
sat_sex1 = sat[sat['sex'] == 1]
sat_sex0 = sat[sat['sex'] == 0]
# Set the number of pairs
num_pairs = 100

# Create an empty dataframe to store the pairs
ps = pd.DataFrame(columns=sat.columns)

# Loop through a range of the desired number of pairs and randomly select one row with sex=1 and one row with sex=0 for each iteration
i = 0
while i<num_pairs:
    # Randomly select a row with sex=1
    pair_sex1 = sat_sex1.sample(n=1)
    
    # Check if there are any rows in sat_sex0 with the same fy_gpa value as the randomly selected row from sat_sex1
    possible_pairs_sex0 = sat_sex0[sat_sex0['fy_gpa'] == pair_sex1['fy_gpa'].values[0]]
    if len(possible_pairs_sex0) == 0:
        continue
    
    # Randomly select a row with sex=0 from the filtered sat_sex0 dataframe
    pair_sex0 = possible_pairs_sex0.sample(n=1)
    
    # Concatenate the two selected rows into a single row and append to the pairs dataframe
    ps = ps.append(pair_sex0, ignore_index = True)
    ps = ps.append(pair_sex1, ignore_index = True)
    
    
    #drop to not sample again
    sat_sex0 = sat_sex0.drop(pair_sex0.index)
    sat_sex1 = sat_sex1.drop(pair_sex1.index)

    #print(len(sat_sex0))

    i += 1 

# Print the resulting pairs dataframe
#print(pairs)
```

```python
sat = ps.reset_index()
```

```python
# create an empty dictionary to store the pairs for each column
d = {}

# iterate over each column
for ind,col in enumerate(['sat_v', 'sat_m', 'hs_gpa']):
    
    # create an empty list to store the pairs for the current column
    pairs = set()
    
    # iterate over all possible pairs of rows
    for i in range(len(sat)):
        for j in range(i+1, len(sat)):
            
            # check if the score of the current column for row i is greater than that of row j
            if sat.loc[i, col] > sat.loc[j, col]:
                pairs.add((i, j))
            
            # check if the score of the current column for row j is greater than that of row i
            elif sat.loc[j, col] > sat.loc[i, col]:
                pairs.add((j, i))
    
    # add the list of pairs to the dictionary for the current column
    d[ind] = pairs
```

```python
d
```

```python
classes = sat['sex']
SAT_pc = opt_fair.Pairwise_with_rev(data = d, penalty = 0, classes = classes)
```

```python
all_pc_sat  = opt_fair._pc_without_reviewers(d)
size = len(sat)
```

```python
%%time
estim = choix.opt_pairwise(size ,all_pc_sat, alpha=0, method='Newton-CG', initial_params=None, max_iter=None, tol=1e-05)
```

```python
stats.kendalltau(estim,sat['fy_gpa'])
```

```python
%%time
A = opt_fair._matrix_of_comparisons(size,all_pc_sat)
P = opt_fair._trans_prob(A)
rank_centrality_temp = np.log(opt_fair._stationary_dist(P))
stats.kendalltau(estim,rank_centrality_temp)
```

```python
%%time
num_reviewers = 3
annot_bt_temp,annot_bias =  opt_fair._alternate_optim(size, num_reviewers, SAT_pc, iters = 11)
```

```python
stats.kendalltau(annot_bt_temp,sat['fy_gpa'])
```
