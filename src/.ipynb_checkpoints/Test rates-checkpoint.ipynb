```python
import numpy as np
import choix
import opt_fair
from scipy.optimize import minimize
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
from matplotlib import colors
import seaborn as sns
```

```python
size = 100 #number of items
min_size = 30 
maj_size = size - min_size
num_reviewers = 50 #number of reviewers
pair_per_reviewer = 300 #number of pairs evaluated per reviewer
```

```python
#init original scores
np.random.seed(42)
maxv=   5
minv=  -5

original = np.random.uniform(minv,maxv ,size = size)
#original[0:(maj_size)] = original[0:(maj_size)] - np.mean(original[0:(maj_size)])
#original[(maj_size):size] = original[(maj_size):size] - np.mean(original[(maj_size):size])
#init classes
classes = [0] * maj_size + [1] * min_size
```

```python
def _alt_alternate_optim(size, num_reviewers, pc_with_revs, original, revs_bias,  iters = 101, tol = 1e-5, gtol = 1e-5):
    '''x0 is the estimated scores
       y0 is the estimated bias for each reviewer'''
    x0 = np.zeros(size) #initial values
    y0 = np.zeros(num_reviewers) #initial values
    obj = []
    k_tau = []
    revs_mse = []
    

    for i in range(iters):

        # minimize with x fixed and update y
        res_y = minimize(lambda y: pc_with_revs.objective(x0, y), y0,tol = tol, jac=lambda y: pc_with_revs.gradient_revs(x0, y), options={"gtol": gtol,'maxiter': 1})
        y0 = res_y.x

        obj_tmp = pc_with_revs.objective(x0, y0)
        obj.append(obj_tmp)
        
        # minimize with y fixed and update x
        res_x = minimize(lambda x: pc_with_revs.objective(x, y0), x0,tol = tol, jac=lambda x: pc_with_revs.gradient_scores(x, y0), options={"gtol": gtol,'maxiter': 1})
        x0 = res_x.x

        obj_tmp = pc_with_revs.objective(x0, y0)
        k_tau_tmp,tmp = stats.kendalltau(x0,original)
        revs_mse_tmp = np.square(np.subtract(y0, [-x for x in revs_bias])).mean()
        
        obj.append(obj_tmp)
        k_tau.append(k_tau_tmp)
        revs_mse.append(revs_mse_tmp)
        
        if ((i) % 20 == 0):
            print(f"Iteration {i}: kendall_tau = {k_tau_tmp}, reviewers bias estimation = {revs_mse_tmp}")
            print(res_x.success)
            print(res_y.success)
            print(obj_tmp)

        if res_x.success and res_y.success:
            print("Minimum found!")
            print(f"Iteration {i}: x = {x0}, y = {y0}")
            #print(res_x.success)
            #print(res_y.success)
            break
    return x0,y0,k_tau,revs_mse,obj
```

```python
%%time
np.random.seed(43)


revs_bias = np.random.uniform(-5, 5, size=num_reviewers)

biased_scores = opt_fair._create_matrix_biased_scores(original,revs_bias,classes)
revs_set =  opt_fair._create_pc_set_for_reviewers(biased_scores,pair_per_reviewer)

pc_with_revs = opt_fair.Pairwise_with_rev(data = revs_set, penalty = 0, classes = classes)
annot_bt,annot_bias,k_tau,revs_mse,obj =  _alt_alternate_optim(size, num_reviewers, pc_with_revs,  original, revs_bias, iters = 1000)

```

```python
data_likel = pc_with_revs.objective(original, -revs_bias)
obj_plt = obj[:]
plt.plot(np.arange(0, len(obj_plt)*0.5, 0.5), [q - data_likel for q in obj_plt])
plt.xlabel('Iterations')
plt.ylabel('Negative LogLikelihood')
plt.title('Convergence of Negative LogLikelihood')
plt.show()
```
