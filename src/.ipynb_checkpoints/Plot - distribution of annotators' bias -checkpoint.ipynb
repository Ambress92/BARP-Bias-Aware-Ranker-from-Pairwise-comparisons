```python
import numpy as np
import choix
import opt_fair
from scipy.optimize import minimize
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
from matplotlib import colors
```

```python
size = 100 #number of items
min_size = 30 
maj_size = size - min_size
num_reviewers = 20 #number of reviewers
pair_per_reviewer = 200 #number of pairs evaluated per reviewer
```

```python
#init original scores
np.random.seed(42)
maxv=   10
minv=  -10
original = np.random.uniform(minv,maxv ,size = size); #print(original)

#original[0:(maj_size)] = original[0:(maj_size)] - np.mean(original[0:(maj_size)])
#original[(maj_size):size] = original[(maj_size):size] - np.mean(original[(maj_size):size])
#init classes
classes = [0] * maj_size + [1] * min_size
```

```python
%%time
np.random.seed(43)
fraction_biased_revs = [i/10 for i in [1,2,3,5,10]]
sd = np.std(original)
bias_fraction =  [i/10 for i in [1,3,5,10]]
bias_vals = [sd*i for i in bias_fraction]
annot_bias = np.zeros((len(bias_vals), len(fraction_biased_revs)))


for i,bias in enumerate(bias_vals):
    for j,frac in enumerate(fraction_biased_revs):
        biased_revs = int(frac * num_reviewers)
        unbiased_revs = num_reviewers - biased_revs
        revs_bias = [bias]*biased_revs + [0]*unbiased_revs
        
        biased_scores = opt_fair._create_matrix_biased_scores(original,revs_bias,classes)
        revs_set =  opt_fair._create_pc_set_for_reviewers(biased_scores,pair_per_reviewer)

        pc_with_revs = opt_fair.Pairwise_with_rev(data = revs_set, penalty = 0, classes = classes)
        annot_bt_temp,annot_bias_temp =  opt_fair._alternate_optim(size, num_reviewers, pc_with_revs, iters = 500)
        
        annot_bias[i,j] = np.square(np.subtract(annot_bias_temp, [-x for x in revs_bias])).mean()
        
        print(frac, bias_fraction[i],bias, annot_bias[i,j])
```

```python
for i,bb in enumerate(bias_fraction):
    plt.plot(fraction_biased_revs, annot_bias[i,:], label = bb)


# Add axis labels
plt.xlabel('Fraction of biased annotators')
plt.ylabel('Mean average error')

# Add a legend
plt.legend()

# Add a title
plt.title('Methods comparison for annotators with different biases')

# Display the plot
plt.show()
```
